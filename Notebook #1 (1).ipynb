```python
#1. Python function: Input GCT file --> Output Pandas Dataframe & optional summary parameter to output # of rows and columns
#function preform manipulations for any data so don't have to repeat same code over and over again 
import pandas as pd

#Set boolean summary to default of false
summary = False

def generate_df(file, summary_info):
    #Local df
    df = pd.read_csv(file)
    
    #If summary is true, print the number of rows and columns
    if summary_info == True: 
        rows = len(df.axes[0])
        cols = len(df.axes[1])
        print("Number of Rows: " + str(rows))
        print("Number of Columns: " + str(cols))
        
    #Print dataframe 
    print(df)
    
    return df
```

```python
#2a. Use the function you wrote in #1 to read this file and print the summary:  
summary = True

generate_df('2aFile.csv', summary)
```

```python
#2b. Use that same function to read this file (without setting a value of the parameter summary in your function call):
df_2b = generate_df('2bFile.csv', summary) 
```

```python
from matplotlib import pyplot as plt
#3. Using the data in the DataFrame from #2b, plot a histogram for each of the following 4 columns

#Get each column individually
column1 = df_2b['A7-A0DB-normal']
column2 = df_2b['A7-A13E-normal']
column3 = df_2b['BH-A0B3-primary']
column4 = df_2b['BH-A0B5-primary']

#Set bin to range from 10 to 10,000 with increments of 1000
bins = range(10, 10000, 1000)

plt.xlabel('X-Axis: Sample Value')
plt.ylabel('Y-Axis: Frequency of Value') 

#Convert column into a list of values 
col1_list = column1.values.tolist()
#Set sample_values as x values
plt.hist(col1_list, bins=bins, edgecolor='black')
plt.title('A7-A0DB-normal\n\n',fontweight ="bold")
plt.show()

#Same code for Column 2, 3, & 4
col2_list = column2.values.tolist()
plt.hist(col2_list, bins=bins, edgecolor='black')
plt.title('A7-A13E-normal\n\n',fontweight ="bold")
plt.show()

col3_list = column3.values.tolist()
plt.hist(col3_list, bins=bins, edgecolor='black')
plt.title('BH-A0B3-primary\n\n',fontweight ="bold")
plt.show()

col4_list = column4.values.tolist()
plt.hist(col4_list, bins=bins, edgecolor='black')
plt.title('BH-A0B5-primary\n\n',fontweight ="bold")
plt.show()
```

```python
#4. Using the data in the DataFrame from #2b, print out a DataFrame containing one row for each samplewith the following columns:
#Sample Name, Mean, Median, Standard Deviation

from IPython.display import display, HTML
import statistics as stat
import numpy as np

#Create DataFrame
dict = {'Sample Name':[],
        'Mean':[],
        'Median':[],
        'Std Deviation':[]
       }
df = pd.DataFrame(dict)

sample_names = ['A7-A0CE-normal','A7-A0CH-normal','A7-A0D9-normal','A7-A0DB-normal','A7-A13E-normal','A7-A13F-normal','A7-A13G-normal','AC-A23H-normal','AC-A2FB-normal', 'AC-A2FF-normal', 'AC-A2FM-normal', 'BH-A0AU-normal', 'BH-A0AY-normal', 'BH-A0AZ-normal', 'BH-A0B3-normal', 'BH-A0B5-normal', 'BH-A0B7-normal', 'BH-A0B8-normal', 'BH-A0BA-normal', 'BH-A0Bc-normal', 'A7-A0CE-primary', 'A7-A0CH-primary', 'A7-A0D9-primary', 'A7-A0DB-primary', 'A7-A13E-primary', 'A7-A13F-primary', 'A7-A13G-primary', 'AC-A23H-primary', 'AC-A2FB-primary', 'AC-A2FF-primary','AC-A2FM-primary', 'BH-A0AU-primary','BH-A0AY-primary','BH-A0AZ-primary','BH-A0B3-primary','BH-A0B5-primary','BH-A0B7-primary','BH-A0B8-primary','BH-A0BA-primary','BH-A0BC-primary']

#Iterate through samples
for name in sample_names:
    
    #Find sample name 
    sample_name = name 
    column = df_2b[name]
    
    #Convert sample data to a list 
    column_list = column.values.tolist()
    
    #Find mean
    mean = sum(column_list)/len(column_list)
    
    #Find median
    median = stat.median(column_list)
    
    #Find standard deviation
    std_deviation = stat.stdev(column_list)
    
    #Compile computed values in a row
    new_row = {'Sample Name': sample_name, 'Mean': mean, 'Median': median, 'Std Deviation': std_deviation}
    
    #Append entire row to table 
    df = df.append(new_row, ignore_index = True)

display(df)
```

```python
#5a. Using the data in the DataFrame from #2b, filter out (remove) the rows whose maximum value is less than 500

#Create new dataframe with rows whose values are greater than 500 
new_df = df_2b 

#Find max of each row 
row_max_series = new_df.max(axis=1)

#Iterate through max in each row 
for indx, value in row_max_series.items():
    
    #If max is less than 500, drop the row 
    if value < 500:
        new_df = new_df.drop(indx)
        
print(new_df)
```

```python
# 5b. What are the mean and median of each column after applying this filter? 

dict = {'Sample Name':[],
        'Mean':[],
        'Median':[]
       }
table = pd.DataFrame(dict)

for name in sample_names:
    sample_name = name 
    
    #Find mean
    mean = sum(new_df[name])/len(new_df[name])

    #Find median
    median = stat.median(new_df[name])

    #Compile computed values in a row
    new_row = {'Sample Name': sample_name, 'Mean': mean, 'Median': median}

    #Append entire row to table 
    table = table.append(new_row, ignore_index = True)

display(table)
```
